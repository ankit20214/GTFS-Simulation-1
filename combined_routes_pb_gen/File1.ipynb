```python
import pandas as pd
trips = open('trips.txt','r')
trips_data = trips.readlines()
trips.close()
route_info = {}
for i in range(len(trips_data)):
    trips_data[i] = trips_data[i].strip()
    trips_data[i] = trips_data[i].split(',')
```

```python
# route_id,service_id,trip_id,trip_headsign,trip_short_name,direction_id,block_id,shape_id,wheelchair_accessible,bikes_allowed
trip_to_route_mapping = {}
for i in range(1,len(trips_data)):
    trip_to_route_mapping[trips_data[i][2]] = trips_data[i][0]
trip_to_route_mapping
```

```python
route_id_to_trip_list = {}
for i in range(1,len(trips_data)):
    if trips_data[i][0] not in route_id_to_trip_list.keys():
        route_id_to_trip_list[trips_data[i][0]] = [trips_data[i][2]]
    else:
        route_id_to_trip_list[trips_data[i][0]] += [trips_data[i][2]]
    
        
```

```python
uniq_list = {}
for k in route_id_to_trip_list.keys():
    uniq_list[int(k)] = int(route_id_to_trip_list[k][0])
```

```python
df= pd.DataFrame()
df['route_id'] = [k for k in uniq_list.keys()]
df['trip_id'] = [v for v in uniq_list.values()]
df.sort_values(by = ['route_id'])
```

```python
trips = open('trips.txt','r')
trips_data = trips.readlines()
trips.close()
shapes_dict = {}
for i in range(1,len(trips_data)):
    trips_data[i] = trips_data[i].split(",")
    shapes_dict[trips_data[i][0]] = trips_data[i][-3]
shapes_dict
```

```python
# Trip id to route id
# Route information , segregate up or down for each line.
# Calculate the average speed of each route
stop_time = open('stop_times.txt','r')
stop_times_data = stop_time.readlines()
stop_time.close()
stop_ids = []
index = []
for i in range(1, len(stop_times_data)):
    stop_times_data[i] = stop_times_data[i].split(',')
    if len(stop_ids) == 0:
        stop_ids += [stop_times_data[i][0]]
        index += [1]
    else:
        if stop_times_data[i][0] != stop_ids[-1]:
            stop_ids += [stop_times_data[i][0]]
            index += [i]
index += [len(stop_times_data)]
```

```python
st = open('stop_times.txt','r')
std = st.readlines()
st.close()
```

```python
file = open('unique_dist.txt','w+')
trips = [v for v in uniq_list.values()]
for t in trips:
    sti = index[t]
    eni = index[t+1]
    tmp = std[sti:eni]
    for tm in tmp:
        file.write(tm)
```

```python
def seconds(current_time):
    return current_time.second + current_time.minute*60 + current_time.hour*3600
import datetime
newf= open('newuniq.txt','r')
new_d = newf.readlines()
newf.close()
ndup = new_d[:]
ndup = [ndup[i].split(',') for i in range(len(ndup))]
for i in range(len(ndup)-1):
    if ndup[i][0] == ndup[i+1][0]:
        cdt = ndup[i][2]
        nat = ndup[i+1][1]
        cdt = datetime.datetime.strptime(cdt, '%H:%M:%S')
        nat = datetime.datetime.strptime(nat, '%H:%M:%S')
        time = seconds(nat) - seconds(cdt)
        dist = float(ndup[i+1][8]) - float(ndup[i][8])
        avg = dist/time
        avg = "{:.3f}".format(round(avg, 3))
        print(avg)
    else:
        avg = 0
    new_d[i] = new_d[i].strip()
    new_d[i] = new_d[i] + "," + str(avg) + "\n"
new_d[-1] = new_d[-1].strip()
new_d[-1] = new_d[-1] + "," + "0" + "\n"
f = open('last_final.txt','w+')
for i in new_d:
    f.write(i)
f.close()
    
```
